# Extract sample data from XQ-1901

Date: 2018-01-01 to 2018-12-31

Datasets: `cube.rb`

```r
ld(r.cube.rb.mst.1901)
```

```r
cube.rb <- r.cube.rb.mst.1901[created.at %between% c(as.Date('2018-01-01'), as.Date('2018-12-31')), .(cube.symbol, stock.symbol, price, created.at, prev.weight.adjusted, target.weight, cube.type)
    ][is.na(prev.weight.adjusted), prev.weight.adjusted := 0
    ] %>% na.omit() %>% unique()
```

```r
sv(cube.rb, 'DM')
```

Datesets: `user.info`

```r
ld(r.cube.info.mst.1901)
```

```r
r.cube.info.mst.1901[1:5]
```

```r
cube.info <- r.cube.info.mst.1901[, .(cube.symbol, owner.id, market, create.date, fans.count)] %>% unique()

sv(cube.info, 'DM')
```

Datesets: `user.stock`

```r
ld(r.user.stock.mst.1901)
```

```r
r.user.stock.mst.1901[1:5]
```

```r
user.stock <- r.user.stock.mst.1901[createAt != 0, .(user.id, code, createAt = as.POSIXct(createAt/1000, origin="1970-01-01"), exchange)]
```

```r
sv(user.stock, 'DM')
```
